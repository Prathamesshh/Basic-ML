```python
# Convert to datetime, automatically handles many formats
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')  # Coerce invalid formats to NaT

# Extract components
df['date_of_journey'] = df['date'].dt.day
df['month_of_journey'] = df['date'].dt.month
df['year_of_journey'] = df['date'].dt.year
```
