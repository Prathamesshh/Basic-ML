```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
plt.title('Sample Plot')
plt.show()
```

```python
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 16, 25]

plt.plot(x, y,color='red', marker='o', linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
```

```python
#multiple plots
x1=[1, 2, 3, 4, 5]
y1=[1, 4, 9, 16, 25]
y2=[2, 3, 5, 7, 11]
x2=[1, 2, 3, 4, 5]
plt.title('Sample Plot')
plt.plot(x1,y1,label="1st",color='red')
plt.plot(x2,y2,label="2nd",color='blue')
```

```python
#bar plot
category=['A', 'B', 'C', 'D', 'E']
y=[1, 4, 9, 16, 25]
plt.bar(category, y,color='violet')
```

```python
#histogram
category=['A', 'B', 'C', 'D', 'E']
y=[1, 4, 9, 16, 25]
plt.hist(y, bins=5, color='green',edgecolor='violet')
```

```python
#scatter plot
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 6, 5]
plt.scatter(x, y, color='blue', marker='x')
```

```python
#pie chart
labels = ['A', 'B', 'C', 'D'] 
sizes = [15,     30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
```
