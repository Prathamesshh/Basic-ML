```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
plt.title('Sample Plot')
plt.show()
```

```python
x=[1, 2, 3, 4, 5]
y=[1, 4, 9, 16, 25]

plt.plot(x, y,color='red', marker='o', linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')    
```
